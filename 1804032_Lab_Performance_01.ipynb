```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/store_data.csv',header=None)
df.head()
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
TID = []

for i in range(0,len(df)):
    TID.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
```

```python
TID[0]
```

```python
rules = apriori(TID, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2)
```

```python
expected_result = list(rules)
expected_result
```

```python
df_results = pd.DataFrame(expected_result)
```

```python
df_results.head()
```

```python
Support = df_results.support
```

```python
first_value = []
second_value = []
third_value = []
fourth_value = []

for i in range(df_results.shape[0]):
    single_list = df_results['ordered_statistics'][i][0]
    first_value.append(list(single_list[0]))
    second_value.append(list(single_list[1]))
    third_value.append(single_list[2])
    fourth_value.append(single_list[3])
```

```python
L = pd.DataFrame(first_value)
R = pd.DataFrame(second_value)

confidence = pd.DataFrame(third_value,columns=['Confidence'])

lift = pd.DataFrame(fourth_value,columns=['Lift'])
```

```python
# concat all list together in a single dataframe
df_final = pd.concat([L, R, Support, confidence, lift], axis=1)
df_final
```

```python
df_final.fillna(value=' ', inplace=True)
df_final.head()
```

```python
df_final.columns = ['L', 1, 'R', 2, 3, 'Support','Confidence','Lift']
df_final.head()
```

```python
df_final['L'] = df_final['L'] + str(", ") + df_final[1]

df_final['R'] = df_final['R']+str(", ") + df_final[2] + str(", ") + df_final[3]
```

```python
df_final.head()
```

```python
df_final.drop(columns=[1,2,3],inplace=True)
```
